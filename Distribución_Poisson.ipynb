```python
#Importacion de librerias 
import scipy.stats as ss # Es para analizar la distribución normal y calcular sus diferentes valores de funcion de distribución
import matplotlib.pyplot as plt #Es una libreria que es orientada a la creación de visualización dinámicas o estáticas de datos
import numpy as np #Es una libreria que es especializada en el cálculo numérico y el análisis de datos, especialmente para un gran volumen de datos.
from numpy import random # Se  un generador de arrgelo de forma especificada con valores aleatorios
import seaborn as sns #Se utiliza para el trazado estadístico en Python.
```

```python
#Nota se tiene que correr uno por uno a partir de este punto.
```

```python
A=ss.poisson(0.1) #Metiendo la variable es A y con el argumento que es 0.1
b=A.cdf(5) #Se metio el comando cdf para una funcion de y=f(x) de 5
print(b) #Se imprime el resultado
```

```python
Y=ss.poisson(6) #Metiendo la variable es Y y con el argumento es de 6
c=Y.sf(10) #Se metio el comando sf que es igual a 1- cdf(x)
print(c) #Se imprime el resultado
```

```python
#Grafica de la Distribución de Poisson
y=np.arange(30) #Es la separacion de las lineas de la grafica 
plt.plot(y,Y.pmf(y),"ro") # Se usa para la creación de graficos en dos dimensiones
plt.vlines(y,0,Y.pmf(y),"y") #Esta linea es para que graficara lineas 
plt.title("Distribución de Poisson") #El titulo de tabla 
plt.ylabel("Probabilidad") #En esta linea se pone el titulo en vertical
plt.xlabel("Valores") #En esta linea se pone el titulo en horizontal
plt.show() #Es una interfaz basada en estados para unmódulo Matplotlib que proporciona una interfaz similar a MATLAB.
```
